**Fetch Employment Data for All Counties in New York**

```python
import requests
import pandas as pd


# Base URL for the ACS data
BASE_URL = 'https://api.census.gov/data/2022/acs/acs5'

# Parameters for the API request
params = {
    'get': 'NAME,B23025_003E,B23025_005E',  # County name, employed, unemployed
    'for': 'county:*',
    'in': 'state:36',  # 36 is the FIPS code for New York
    'key': API_KEY
}

# Function to fetch employment data for all counties in NY
def fetch_employment_data():
    try:
        response = requests.get(BASE_URL, params=params)
        response.raise_for_status()
        data = response.json()
        df = pd.DataFrame(data[1:], columns=data[0])
        return df
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None

# Fetch the data
employment_data_df = fetch_employment_data()
if employment_data_df is not None:
    # Display the first few rows of the data
    print("Employment Data for NY Counties:")
    print(employment_data_df.head())
else:
    print("Failed to fetch data.")
```
